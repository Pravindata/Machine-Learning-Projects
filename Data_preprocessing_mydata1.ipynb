```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris_data=load_iris()
```

```python
data=pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
```

```python
print("Skewness of Sepal l : ",data["sepal length (cm)"].skew())
print("Skewness of Sepal W : ",data["sepal width (cm)"].skew())
print("Skewness of Petal l : ",data["petal length (cm)"].skew())
print("Skewness of Petal w : ",data["petal width (cm)"].skew())
```

```python
print("Skewness of Sepal l : ",data["sepal length (cm)"].skew())
print("Skewness of Sepal l : ",sns.distplot(data["sepal length (cm)"]))
```

```python
print("Skewness of Sepal W : ",data["sepal width (cm)"].skew())
print("Skewness of Sepal w : ",sns.distplot(data["sepal width (cm)"]))
```

```python
print("Skewness of Petal l : ",data["petal length (cm)"].skew())
print("Skewness of Petal l : ",sns.distplot(data["petal length (cm)"]))
```

```python
print("Skewness of Petal w : ",data["petal width (cm)"].skew())
print("Skewness of Petal w : ",sns.distplot(data["petal width (cm)"]))
```

```python
print("Kurtosis of Sepal l : ",data["sepal length (cm)"].kurt())
print("Kurtosis of Sepal W : ",data["sepal width (cm)"].kurt())
print("Kurtosis of Petal l : ",data["petal length (cm)"].kurt())
print("Kurtosis of Petal w : ",data["petal width (cm)"].kurt())
```
